```python
#one hot encoding

data_dc = pd.get_dummies(data_dc, columns=['STYLE','STRUCT','EXTWALL','ROOF','INTWALL','ASSESSMENT_NBHD','HEAT'])
```

```python
def AC_numeric(value):
    if (value == 'Y'):
        return int('1') 
    else:
        return int('0')

def QUALIFIED_numeric(value):
    if (value == 'Q'):
        return int('1') 
    else:
        return int('0')

data_dc['AC_numeric'] = data_dc['AC'].apply(AC_numeric)
data_dc = data_dc.drop(columns='AC')

data_dc['QUALIFIED_numeric'] = data_dc['QUALIFIED'].apply(QUALIFIED_numeric)
data_dc = data_dc.drop(columns='QUALIFIED')
```

```python
def GradeToNumeric(value):
        if (value == 'Exceptional-D'):
            return int(12)
        elif (value == 'Exceptional-C'):
            return int(11)
        elif (value == 'Exceptional-B'):
            return int(10)
        elif (value == 'Exceptional-A'):
            return int(9)
        elif (value == 'Superior'):
            return int(8)
        elif (value == 'Excellent'):
            return int(7)
        elif (value == 'Very Good'):
            return int(6)
        elif (value == 'Good Quality'):
            return int(5)
        elif (value == 'Good Average'):
            return int(4)
        elif (value == 'Above Average'):
            return int(3)
        elif (value == 'Average'):
            return int(2)
        elif (value == 'Fair Quality'):
            return int(1)
        else:
            return int(2)

data_dc['GRADE_numeric'] =data_dc['GRADE'].apply(GradeToNumeric)
data_dc = data_dc.drop(columns='GRADE')
```

```python
def CNDTNToNumeric(value):
        if (value == 'Excellent'):
            return int(6)
        elif (value == 'Very Good'):
            return int(5)
        elif (value == 'Good'):
            return int(4)
        elif (value == 'Average'):
            return int(3)
        elif (value == 'Fair'):
            return int(2)
        elif (value == 'Poor'):
            return int(1)
        elif (value == 'Default'):
            return int(3)

data_dc['CNDTN_numeric'] =data_dc['CNDTN'].apply(CNDTNToNumeric)
data_dc = data_dc.drop(columns='CNDTN')
```

```python
data_dc.info()
```

```python
#div dataset to test and train data 
all_column = list(data_dc.columns.values)
X = data_dc[all_column]
y = data_dc['PRICE']

train_dataset, test_dataset = train_test_split(data_dc, test_size=0.2)

y_train = train_dataset.PRICE.values

x_train = train_dataset.drop('PRICE',axis =1)

y_test = test_dataset.PRICE.values

x_test = test_dataset.drop('PRICE',axis=1)
```

```python
# Linearna regresia
l_model = linear_model.LinearRegression()
l_model.fit(x_train,y_train)

l_y_pred = l_model.predict(x_test)
plt.scatter(y_test,l_y_pred)
#ak by predikovane hodnoty boli rovnake ako y_test hondoty (respektive skutocne hodnoty), bolo by rozlozenie bodov na grafe nizsie v osi uhla xy
```

```python
print("Predikcia cien:", l_y_pred)  
rmse_l = numpy.sqrt(mean_squared_error(l_y_pred, y_test))

print("Root mean square error je ", rmse_l)

print("R2 skore je ",r2_score(y_test,l_y_pred))
```

```python
dtr_model = DecisionTreeRegressor(random_state = 0)
dtr_model.fit(x_train,y_train)
dtr_y_pred = dtr_model.predict(x_test)
  
print("Predikcia cien:", dtr_y_pred)  
dtr_rmse = numpy.sqrt(mean_squared_error(dtr_y_pred, y_test))

print("Root Mean square error je ", dtr_rmse)

print("R2 skore je ",r2_score(y_test,dtr_y_pred))
```
